```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')



# Function to build the model
def build_model(input_shape, out_shape, activ_str, neurons_layers):
    model = Sequential()
    model.add(layers.InputLayer(input_shape=input_shape))

    for neurons in neurons_layers:
        model.add(layers.Dense(neurons))
        model.add(layers.Activation(activ_str))
        model.add(layers.Dropout(0.2))

    model.add(layers.Dense(out_shape, activation="sigmoid"))
    return model

# Main function
def main():
    # Load dataset
    df = pd.read_csv('/content/drive/My Drive/beacon_readings.csv', encoding='latin1')


    neuron_layers = [[128, 64], [128, 64, 32], [128, 64, 32, 16], [128, 64, 32, 16, 8]]
    activ_str = ["sigmoid", "relu", "tanh"]
    norm_factors = np.array([274., 438.])
    norm_fac = np.linalg.norm(norm_factors)
    beacon_position = np.array([[0.10, 1.80], [2.74, 2.66], [1.22, 4.38]]) * 100.

    X = df.drop(["Position X", "Position Y", "Date", "Time"], axis=1)
    y = df.drop(["Distance A", "Distance B", "Distance C", "Date", "Time"], axis=1)
    X = np.array(X)
    y = np.array(y)
    y_normalized = y / norm_fac

    true_distance = []
    for i in range(y.shape[0]):
        t = y[i]
        d = np.linalg.norm(beacon_position - t, ord=2, axis=1)
        true_distance.append(d)
    true_distance = np.array(true_distance) / norm_fac

    add_u = True
    if add_u:
        u = np.zeros((X.shape[0], 1)) + (-12)
        X = np.concatenate([X, u], axis=1)

    X_train, X_test, y_train, y_test = train_test_split(X, true_distance, test_size=0.2)

    z = 1
    for m, layer_config in enumerate(neuron_layers):
        plt.figure(z)
        for j, activation in enumerate(activ_str):
            model = build_model((X.shape[1],), 3, activation, layer_config)
            model.compile(optimizer="adam", loss="mse")

            r2_array = []
            for i in range(1, 26):
                model.fit(X_train, y_train, batch_size=8, epochs=10, validation_data=(X_test, y_test), verbose=0)
                predictions = model.predict(X_test)
                r2_score = metrics.r2_score(y_test, predictions)
                r2_array.append(r2_score)

            plt.plot(np.arange(1, 26) * 10, r2_array, label=f"{activation}")
            for n, score in enumerate(r2_array):
                plt.text((n + 1) * 10, score, f"{score:.3f}")

        plt.title(f'{len(layer_config)} hidden layers')
        plt.xlabel('Number of iterations')
        plt.ylabel('R2 Score')
        plt.legend()
        z += 1
    plt.show()

# Entry point
if __name__ == "__main__":
    main()

```
